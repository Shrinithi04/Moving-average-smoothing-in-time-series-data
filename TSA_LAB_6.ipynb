```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

```python
def moving_average_smoothing(data, window_size):
    return data.rolling(window=window_size).mean()
```

```python
dates = pd.date_range(start='2024-01-01', periods=50, freq='D')
values = np.cumsum(np.random.randn(50) * 2)
df = pd.DataFrame({'Date': dates, 'Value': values})
df.set_index('Date', inplace=True)
```

```python
window_size = 5
df['Smoothed'] = moving_average_smoothing(df['Value'], window_size)
```

```python
print("Original Data:")
print(df['Value'])
print("\nData after Applying Moving Average Smoothing:")
print(df['Smoothed'])
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df.index, df['Value'], label='Original Data', linestyle='dashed')
plt.plot(df.index, df['Smoothed'], label=f'{window_size}-Day Moving Average', linewidth=2)
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Moving Average Smoothing for Time Series Data')
plt.legend()
plt.grid()
plt.show()
```
